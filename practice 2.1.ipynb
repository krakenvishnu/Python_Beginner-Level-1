# **Variables**

Syntax

```python
variable_name = value
```

```python
# example 1: Integer Variable
age =25 # here, 'age' is a variable, storing an integer value 25
print(age) # output: 25
```

```python
# example 2: String Variable
name = "John" # here, 'name' is a variable, storing a string value "John"
print(name) # output:"John"
```

```python
# example 3: Float Variable
price = 19.99 #'price' is a variable, storing a float value 19.99
print(price) # output: 19.99
```

```python
# example 4: Boolean Variable
is_active = True #'is_active' is a variable, storing a boolean value True
print(is_active) # output: True
```

1. Storing and Printing a value

```python
# sorting the value 10 in a variable called 'x'
x=10
#printing the value of 'x'
print(x) #OUTPUT: 10
```

2. Using Variables in expressions

```python
# Assigning values to variables
a=5
b=10
c=a+b
print(c) #OUTPUT: 15
```

3. Changing the value of a variable

```python
# initial value
x=5
print(x) #OUTPUT: 5
# changing the value
x=10
print(x) #OUTPUT: 10
```

4. Concatenating Strings

```python
# Assigning values to variables
first_name = "John"
last_name = "Doe"
# Concatenating strings and storing in a new variable
full_name = first_name + " " + last_name
print(full_name) #OUTPUT: John Doe
```

5. Using Variables in a Concatenation

```python
# Assigning values to variables
length = 10
width = 5
# Concatenating strings and variables
area = "The area is " + str(length * width)
print(area) #OUTPUT: The area is 50
```

Re-assigning Values to variables.


python allows you to reassign a varible to a new value at anytime.

```python
x=10
print(x) #OUTPUT: 10
# reassigning the value of 'x'
x=20
print(x) #OUTPUT: 20
```
