**TUPLE**

tuple is an immutable datastructure

it is defined with '(   )'

```python
t=(1,2,3,4,5,6,7,8,9,0)
```

```python
t
```

```python
type(t)
```

```python
t.count(5)
```

```python
t.index(7)
```

tuple don't have append function because it is immutable.

tuple donot function any mutable functions.

```python
t.append(20)
```

```python
for i in (t):
  print(i)
```

```python
for i in enumerate(t):
  print(i)
```
