**Copy Dictionary**

```python
d = {'Name':'Asif' , 'ID': 12345 , 'DOB': 1991 , 'Address' : 'Hilsinki'}
d
```

```python
d8 = d # Create a new reference "d8"
```

```python
d8
```

```python
id(d) , id(d8) # The address of both mydict & mydict1 will be the same
```

```python
d9 = d8.copy() # Create a copy of the dictionary
```

```python
d9
```

```python
id(d9) # The address of d9 will be different from d8 because d9 is an copy of d8
```

```python
d9['Address'] = 'Mumbai'
```

```python
d9
```

** Loop Through a Dictionary **

```python
for i in d9:
  print(i , ':' , d9 [i]) # Key & value pair
```

```python
for i in d9:
  print(d9[i]) # Dictionary items
```
